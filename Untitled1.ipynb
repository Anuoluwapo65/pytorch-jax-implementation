```python
class Activation_function(nn.Module):
  def __init__(self):

    super().__init__()
    self.name = self.__class__.__name__
    self.config = {"name": self.name}


  def forward(self, x):
    pass
```

```python
class Sigmoid(Activation_function):

  def forward(self, x):
    return 1 / ( 1 + torch.exp(-x))

class Tanh(Activation_function):

  def forward(self, x):
    x_pos, x_neg = torch.exp(x), torch.exp(-x)
    return (x_pos - x_neg) / (x_pos + x_neg)
```

```python
class ReLU(Activation_function):

  def forward(self, x):
    return x * (x > 0).float()
```

```python
class LeakyReLU(Activation_function):
  def __init__(self, alpha = 0.1):
    super().__init__()

    self.config["alpha"] = alpha

  def forward(self, x):
    return torch.where(x > 0, x, self.config["alpha"] * x)
```

```python
class ELU(Activation_function):

  def forward(self, x):
    return torch.where(x > 0, x, torch.exp(x) - 1)

class Swish(Activation_function):
  def forward(self, x):

    return x * torch.sigmoid(x)
```

```python
act_fn_name = {
    "sigmoid": Sigmoid,
    "tanh": Tanh,
    "relu": ReLU,
    "leakyrelu":LeakyReLU,
    "elu":ELU,
    "swish":Swish

}

act_fn_name
```

```python
def grad_fn(act_fn, x):

  x = x.clone().requires_grad_()
  out = act_fn(x)
  out.sum().backward()
  return x.grad
```

```python
def visualize_grad(act_fn, ax, x):
  y = act_fn(x)
  y_grad = grad_fn(act_fn, x)
  x, y, y_grad = x.cpu().numpy(), y.cpu().numpy(), y_grad.cpu().numpy()



  ax.plot(x, y, linewidth = 4, label = "Activation_function")
  ax.plot(x, y_grad, linewidth = 4, label = "gradient")
  ax.set_title(act_fn.name)
  ax.legend()
  ax.set_ylim(-1.5, x.max())


act_fns=[act_fn() for act_fn in act_fn_name.values()]
x = torch.linspace(-5, 5, 1000)
rows = math.ceil(len(act_fns)/2.0)
fig, ax = plt.subplots(rows, 2, figsize = (rows * 4, 4))
for i , act_fn in enumerate(act_fns):
    visualize_grad(act_fn, ax[divmod(i, 2)], x)
    fig.subplots_adjust(hspace = 0.5)
plt.show()
```

```python
class Fashion (nn.Module):
  def __init__(self, act_fn, num_input = 784, num_classes = 10, num_hidden = [512, 256, 256, 184]):
    super().__init__()

    layers = []

    layer_size = [num_input] + num_hidden
    for layer_index in range(1, len(layer_size)):
      layers+=[nn.Linear(layer_size[layer_index-1], layer_size[layer_index]),act_fn]
    # Corrected line:
    layers.append(nn.Linear(layer_size[-1], num_classes))
    # Corrected typo: nn.Squential -> nn.Sequential
    self.layers = nn.Sequential(*layers)

    self.config = {"act_fn": act_fn.config, "num_input":num_input, "num_classes":num_classes, "num_hidden":num_hidden}


  def forward(self, x):
    x = x.view(x.size(0), -1)
    x = self.layers(x)
    return x
```

```python
def _config_file(model_path, model_name):
  return os.path.join(model_path, model_name + ".config")
def _model_file(model_path, model_name):
  return os.path.join(model_path, model_name + ".tar")
def load_model(model_name, model_path, net = None):
  config_file, model_file  = _config_file(model_path, model_name), _model_file(model_path, model_name)
  print(f"[Debug Load] Attempting to load config from: {config_file}")
  print(f"[Debug Load] Attempting to load model from: {model_file}")
  assert os.path.isfile(config_file)
  assert os.path.isfile(model_file)

  with open(config_file, "r") as f:
    config_dict = json.load(f)

  if net is None:
    act_fn_nm = config_dict["act_fn"].pop("name").lower()
    act_fn = act_fn_name[act_fn_nm](**config_dict.pop("act_fn"))
    net = Fashion(act_fn = act_fn, **config_dict)

  net.load_state_dict(torch.load(model_file,map_location = device))
  return net


def save_model(model,model_name,model_path):
  config_dict = model.config
  os.makedirs(model_path, exist_ok =True)
  config_file, model_file = _config_file(model_path, model_name),_model_file(model_path, model_name)
  print(f"[Debug Save] Attempting to save config to: {config_file}")
  print(f"[Debug Save] Attempting to save model to: {model_file}")
  with open(config_file, "w")as f:
    json.dump(config_dict, f)
  torch.save(model.state_dict(), model_file)
```

```python
def visualizing_gradient(net, color = "C0"):

  net.eval()


  small_dataloader =  data.DataLoader(train_set, batch_size = 256, shuffle = False, num_workers = 2)
  img, labels = next(iter(small_dataloader))

  img, labels = img.to(device), labels.to(device)
  net.zero_grad()
  pred = net(img)
  pred = pred.squeeze(dim = 1)
  loss = F.cross_entropy(pred, labels)
  loss.backward()
  grads = {name : params.grad.data.view(-1).cpu().numpy() for name, params in net.named_parameters() if "weight" in name}
  net.zero_grad()


  columns = len(grads)
  fig, ax = plt.subplots(1, columns, figsize = (columns * 5, 5.5))
  fig_index = 0
  for key in grads:
    key_ax = ax[fig_index%columns]
    sns.histplot(grads[key], ax =key_ax, bins = 50, color = color, kde = True)
    key_ax.set_title(str(key))
    key_ax.set_xlabel("visualizing Gradient")
    fig_index += 1
  fig.suptitle(f"the visualization of all the graident of all activation function {net.config["act_fn"]["name"]}", fontsize = 15, y = 1.05)
  fig.subplots_adjust(hspace = 0.5)
  plt.show()
  plt.close()
```

```python
def train_model(net, model_name, max_epochs=5, patience=7, batch_size=256, overwrite=False):
    """
    Train a model on the training set of FashionMNIST

    Inputs:
        net - Object of BaseNetwork
        model_name - (str) Name of the model, used for creating the checkpoint names
        max_epochs - Number of epochs we want to (maximally) train for
        patience - If the performance on the validation set has not improved for #patience epochs, we stop training early
        batch_size - Size of batches used in training
        overwrite - Determines how to handle the case when there already exists a checkpoint. If True, it will be overwritten. Otherwise, we skip training.
    """
    file_exists = os.path.isfile(_model_file(checkpoint_path, model_name))
    if file_exists and not overwrite:
        print("Model file already exists. Skipping training...")
    else:
        if file_exists:
            print("Model file exists, but will be overwritten...")

        # Defining optimizer, loss and data loader
        optimizer = optim.SGD(net.parameters(), lr=1e-2, momentum=0.9) # Default parameters, feel free to change
        loss_module = nn.CrossEntropyLoss()
        train_loader_local = data.DataLoader(train_set, batch_size=batch_size, shuffle=True, drop_last=True, pin_memory=True)

        val_scores = []
        best_val_epoch = -1
        for epoch in range(max_epochs):
            ############
            # Training #
            ############
            net.train()
            true_preds, count = 0., 0
            for imgs, labels in tqdm.tqdm(train_loader_local, desc=f"Epoch {epoch+1}", leave=False):
                imgs, labels = imgs.to(device), labels.to(device) # To GPU
                optimizer.zero_grad() # Zero-grad can be placed anywhere before "loss.backward()"
                preds = net(imgs)
                loss = loss_module(preds, labels)
                loss.backward()
                optimizer.step()
                # Record statistics during training
                true_preds += (preds.argmax(dim=-1) == labels).sum()
                count += labels.shape[0]
            train_acc = true_preds / count

         ##############
            # Validation #
            ##############
            val_acc = test_model(net, val_loader)
            val_scores.append(val_acc)
            print(f"[Epoch {epoch+1:2d}] Training accuracy: {train_acc*100.0:05.2f}%, Validation accuracy: {val_acc*100.0:05.2f}%")

            if len(val_scores) == 1 or val_acc > val_scores[best_val_epoch]:
                print("\t   (New best performance, saving model...)")
                save_model(net, checkpoint_path, model_name)
                best_val_epoch = epoch
            elif best_val_epoch <= epoch - patience:
                print(f"Early stopping due to no improvement over the last {patience} epochs")
                break

        # Plot a curve of the validation accuracy
        plt.plot([i for i in range(1,len(val_scores)+1)], val_scores)
        plt.xlabel("Epochs")
        plt.ylabel("Validation accuracy")
        plt.title(f"Validation performance of {model_name}")
        plt.show()
        plt.close()

    load_model(checkpoint_path, model_name, net=net)
    test_acc = test_model(net, test_loader)
    print((f" Test accuracy: {test_acc*100.0:4.2f}% ").center(50, "=")+"\n")
    return test_acc


def test_model(net, data_loader):
    """
    Test a model on a specified dataset.

    Inputs:
        net - Trained model of type BaseNetwork
        data_loader - DataLoader object of the dataset to test on (validation or test)
    """
    net.eval()
    true_preds, count = 0., 0
    for imgs, labels in data_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        with torch.no_grad():
            preds = net(imgs).argmax(dim=-1)
            true_preds += (preds == labels).sum().item()
            count += labels.shape[0]
    test_acc = true_preds / count
    return test_acc
```

```python
for act_by_name in act_fn_name:
    print(f"Training BaseNetwork with {act_fn_name} activation...")
    set_seed(42)
    act_fn = act_fn_name[act_by_name]()
    net_actfn = Fashion(act_fn=act_fn).to(device)
    train_model(net_actfn, f"FashionMNIST_{act_fn_name}", overwrite=False)
```

```python
def activation_gradient(net, color = "C0"):
  activations = {}

  net.eval()
  small_dataloader = data.DataLoader(train_set, shuffle = False, batch_size = 1024)
  data_input, data_label = next(iter(small_dataloader))

  with torch.no_grad():
    layer_index = 0

    data_input = data_input.to(device)
    data_label = data_label.to(device)


    data_input = data_input.view(data_label.size(0),-1)
    for layer_index, layer  in enumerate(net.layers[:-1]):

      data_input = layer(data_input)
      activations[layer_index] = data_input.view(-1).cpu().numpy()
  columns = 4
  rows = math.ceil(len(activations)/ columns)
  fig, ax = plt.subplots(rows, columns , figsize = (rows * 6.5, columns * 7.5))
  fig_index = 0
  for key in activations:
    key_ax = ax[fig_index//columns][fig_index%columns]
    sns.histplot(data =activations[key], ax = key_ax, bins = 50, color = color, kde = True,stat = "density")
    key_ax.set_title(f"Layer {key} - {net.layers[key].__class__.__name__}")
    fig_index += 1
  fig.suptitle(f"Activation distribution for activation function {net.config['act_fn']['name']}", fontsize=14)
  fig.subplots_adjust(hspace=0.4, wspace=0.4)
  plt.show()
  plt.close()


```
